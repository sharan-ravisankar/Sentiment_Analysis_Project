Import all required libraries

```python
import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
import joblib
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')
```

# Create stopword list and lemmatizer

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def full_preprocess(text):
  text=text.str.lower()
  text=text.str.replace(r"\d+", "",regex=True)
  text=text.str.replace(r"http\S+|www\S+|https\S+", "", regex=True)
  text=text.str.translate(str.maketrans("", "", string.punctuation))
  text=text.apply(lambda x: ' '.join([lemmatizer.lemmatize(word)
                                   for word in x.split() if word not in stop_words]))
  text=text.apply(lambda x: ' '.join(re.sub(r'[^\x00-\x7F]+',' ', x).split()))
  text=text.str.strip().str.replace(r"\s+", " ", regex=True)
  return text
```

# Create a preprocessing pipeline to reuse the function easily

```python
preprocess_pipeline = Pipeline([
    ('full_preprocess',FunctionTransformer(full_preprocess,validate=False)),
])
```

#Apply preprocessing to text columns

```python
df['clean_title']=preprocess_pipeline.fit_transform(df['title'])
df['clean_body']=preprocess_pipeline.fit_transform(df['body'])
```

Save preprocessing pipeline

```python
joblib.dump(preprocess_pipeline, "preprocess_pipeline.pkl")
print("Preprocessing pipeline saved as preprocess_pipeline.pkl")
```

# Combine cleaned text columns into one

```python
df['full_text']=df['clean_title']+' '+df['clean_body']
```

```python
df.shape
```

# Intialize sentiment analyzer(VADER)

```python
sia=SentimentIntensityAnalyzer()
```

#Calculate sentiment score for each review

```python
df['sentiment_score']=df['full_text'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
df['sentiment_score']
```

# Convert Sentiment score to label
# > 0.05 -> positive | < -0.05-> Negative | else Neutral

```python
df['sentiment']=df['sentiment_score'].apply(lambda x: 'positive' if x>0.05 else ('negative' if x<-0.05 else 'neutral'))
```

```python
print("sentiment distribution")
print(df['sentiment'].value_counts())
```

```python
plt.figure(figsize=(6,4))
df['sentiment'].value_counts().plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Type')
plt.ylabel('Count')
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler

X = df['full_text']
y = df['sentiment']

# Create oversampler
ros = RandomOverSampler(random_state=42)

# Resample data
X_resampled, y_resampled = ros.fit_resample(X.values.reshape(-1, 1), y)

# Flatten X back to Series
X_resampled = pd.Series(X_resampled.flatten(), name='full_text')
y_resampled = pd.Series(y_resampled, name='sentiment')

print(" After oversampling:")
print(y_resampled.value_counts())
```

```python
import matplotlib.pyplot as plt
y.value_counts().plot(kind='bar', title='Before Oversampling', color=['green','red','gray'])
plt.xlabel('Sentiment Type')
plt.ylabel('Count')
plt.show()

y_resampled.value_counts().plot(kind='bar', title='After Oversampling', color=['green','red','gray'])
plt.xlabel('Sentiment Type')
plt.ylabel('Count')
plt.show()
```

#TF-IDF VECTORISATION

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer=TfidfVectorizer(max_features=5000)
```

```python
tfidf_vectorizer
```

```python
x_tfidf=tfidf_vectorizer.fit_transform(X_resampled)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_tfidf, y_resampled, test_size=0.2, random_state=42)

```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
```

```python
y_pred = nb_model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
